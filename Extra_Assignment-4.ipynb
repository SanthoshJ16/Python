```python
# Passing the return value from the function to another variable
def oddEven():
    num=int(input("Enter the Number:" ))
    if((num%2)==0):
        print(num,"is an Even Number")
        value='This is even Number'
    else:
        print(num,"is an Odd Number")
        value='This is an Odd Number'
    return value
chkValue=oddEven() # Passing the return value to another variable
print("From Function",chkValue) # To show the value inside obj chkValue
```

```python
# Eligibility Check for Marriage
def Eligible():
    gender=input("Enter your Gender:")
    age=int(input("Enter your Age:" ))
    if(gender=='Male' and age>=21):
        eligibilty="Eligible"
    if(gender=='Female' and age>=18):
        eligibilty="Eligible"
    else:
        eligibilty="Not Eligible"
    return eligibilty
```

```python
# calculate the percentage of your 10th mark
def PercentageofMarks(s1,s2,s3,s4,s5):# Used Parameterized function
    total=s1+s2+s3+s4+s5 
    percentage=float(total/5)
    print('Percentage',percentage) #To make sure % is calculated in Function
    return percentage
```

```python
actualPercentage=PercentageofMarks(98,87,95,95,93)
print('Percentage from function',actualPercentage)
```

```python
#print area and perimeter of triangle using class and functions

def areaTriangle(height,breadth):
    print('Height: ',height)
    print('Breadth: ',breadth)
    print('Area formula: (Height*Breadth)/2')
    area=(height*breadth)/2    
    return area
def perimeterTraingle(height1,height2,breadth):
    print('Perimeter formula: Height1+Height2+Breadth')
    print('Height1: ',height1)
    print('Height2: ',height2)
    print('Breadth: ',breadth)
    perimeter=height1+height2+breadth  
    return perimeter
```

```python
area=areaTriangle(32,34)
print('Area of Triangle : ', area)
perimeter=perimeterTraingle(2,4,4)
print('Perimeter of Triangle :', perimeter)
```
